### bertの実行
ここまでの作業の結果で日本語bertが実行可能になります。<br>
Colabの領域でなく、Google driveの領域にカレントフォルダを移動して、
Google drive上のコードを直接実行します。<br>
bertは、pythonの起動時パラメータ指定を処理に使用するように作られているので、<br>
（少しトリッキーですが）jupyterからpythonコマンドを起動する形にしています。

```
# bertのコードフォルダにカレントを移動
%cd /content/drive/My\ Drive/Colab_Code/bert
```

```
%%bash
BERT_BASE_DIR="/content/drive/My\ Drive/Colab_Data/Japanese_L-12_H-768_A-12_E-30_BPE"
python extract_features.py --input_file=./datain/in1.txt --output_file=./dataout/out11j.json \
  --vocab_file="/content/drive/My Drive/Colab_Data/Japanese_L-12_H-768_A-12_E-30_BPE/vocab.txt" \
  --bert_config_file="/content/drive/My Drive/Colab_Data/Japanese_L-12_H-768_A-12_E-30_BPE/bert_config.json" \
  --init_checkpoint="/content/drive/My Drive/Colab_Data/Japanese_L-12_H-768_A-12_E-30_BPE/bert_model.ckpt" \
  --do_lower_case=False --layers=-2 --max_seq_length=128 --batch_size=8
```
